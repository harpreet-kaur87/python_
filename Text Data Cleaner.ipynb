### Text Data Cleaner

##### This project builds a simple Text Data Cleaner in Python that processes a list of raw text strings by removing empty values and normalizing text (lowercasing and stripping spaces). It then calculates the frequency of each word using basic Python data structures like loops and dictionaries.

```python
sentence = ['hello', 'mY', 'HELLO', 'date science', ' ', 'python','sql']

def text_data_cleaner(lst):
    
    word_count = {}

    for item in lst:
        item = item.strip().lower()
        if item != '':
            if item in word_count:
                word_count[item] += 1
            else:
                word_count[item] = 1
                
    return word_count

text_data_cleaner(sentence)
```
