```python
store_inventory = {
            'Milk' : 10,
            'Butter' : 25,
            'Curd' : 12,
            'Bread' : 7,
            'Cheese' : 26,
            'Biscuits' : 32,
            'Cheese Slice' : 19
}

print('Initial Store Inventory\n')
for key,value in store_inventory.items():
    print(f'{key} : {value}')
    
print()
    
def add_product(product,quantity):
    
    if product not in store_inventory:
        store_inventory[product] = quantity
        print(f'{product} has been added to inventory\n')
    else:
        print(f'{product} already exists\n')
        
def update_quantity(product,quan):
    
    if product in store_inventory:
        store_inventory[product] += quan
        print(f'{quan} quantity of {product} has been successfully added\n')
    else:
        print(f'{product} not found\n')
        
def delete_product(product):
    
    if product in store_inventory:
        del store_inventory[product]
        print(f'{product} has been successfully deleted\n')
    else:
        print(f'{product} not found\n')

        
add_product('Sugar',15)
update_quantity('Cheese',10)
delete_product('Butter')
        
print('Latest Inventory : \n')

for key,value in store_inventory.items():
    print(f'{key} : {value}')
```
