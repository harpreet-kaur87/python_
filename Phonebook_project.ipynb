```python
# Simple Contact Book
# Description: Create a program to store and manage contacts using a dictionary where the key is the 
# person's name and the value is their phone number.

# Features to Include:
# Add a new contact
# Search for a contact
# Update a contact
# Delete a contact
# Display all contacts
```

```python
print('Simple Contact Book')
print()
print('Press 1 to add a new contact')
print('Press 2 to update a contact')
print('Press 3 to delete a contact')
print('Press 4 to display all contacts')
print('Press 5 to exit')
print()
print()

contact_book = {'arjun' : '9898987225',
               'harpreet' : '9800000555',
               'sara' : '5555544444'}

while True:
    
    enter_input = input('Enter your option - ')
    print()
    if enter_input == '1':
        enter_name = input('Enter name - ').lower()
        enter_num = input('Enter number - ')
        contact_book[enter_name] = enter_num
        print(f'New contact {enter_name.capitalize()},{enter_num} has been added to contact book.')
    elif enter_input == '2':
        updated_contact_name = input('Enter name - ').lower()
        if updated_contact_name in contact_book:
            updated_contact_num = input('Enter new number - ')
            contact_book[updated_contact_name] = updated_contact_num
            print(f'Contact {updated_contact_name.capitalize()},{updated_contact_num} has been updated.')
        else:
            print(f'Contact not found.')
    elif enter_input == '3':
        delete_contact = input('Enter name - ').lower()
        if delete_contact in contact_book:
            del contact_book[delete_contact]
            print(f'Contact {delete_contact.capitalize()} has been deleted.')
        else:
            print(f'Contact not found.')
    elif enter_input == '4':
        for key,value in contact_book.items():
            print(f'{key.capitalize()}:{value}')
    elif enter_input == '5':
        print(f'You have exited the program')
        break
    else:
        print('Enter valid input')
```
