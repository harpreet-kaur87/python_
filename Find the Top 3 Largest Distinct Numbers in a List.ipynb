#### Write a Python program to find the top 3 largest distinct numbers from a given list without using built-in functions like max(), min(), sort(), or sorted().

```python
lst = [12, 45, 7, 23, 89, 34, 89, 45]

def find_max_values(lst):
    
    lst = list(set(lst))
    for i in range(len(lst)):
        for j in range(i+1, len(lst)):
            if lst[i] > lst[j]:
                lst[i],lst[j] = lst[j],lst[i]
    return lst[-3:][::-1]

solution = find_max_values(lst)
print(solution)
```
