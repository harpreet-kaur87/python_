#### Problem Statement

Write a program to identify out-of-stock items in an inventory system.

The inventory is represented as a dictionary where each key denotes a product category (such as fruits or vegetables), and the corresponding value is a list of items belonging to that category. A separate set contains the names of items that are currently out of stock.

The program should traverse the inventory and determine which items are out of stock, displaying each such item along with its respective category.

```python
inventory = {
    "fruits": ["apple", "banana", "orange"],
    "vegetables": ["carrot", "spinach", "potato"]
}

out_of_stock = {"banana", "potato"}


def find_stockout_items(inventory,out_of_stock):
    
    stockout_inventory = {}
    for category,items in inventory.items():
        stockout_inventory[category] = [item for item in items if item in out_of_stock]
    return stockout_inventory

find_stockout_items(inventory,out_of_stock)
```
