## Student Score Analyzer

##### The Student Score Analyzer is a Python-based mini project that analyzes student academic records. By entering a studentâ€™s name, the program retrieves subject-wise marks and calculates average, highest, and lowest scores using core Python concepts like lists, loops, and conditionals.

```python
student_record = [
    {'name':'Arjun', 'Science' : 75, 'Maths' : 82, 'English' : 62},
    {'name':'Damini', 'Science' : 95, 'Maths' : 100, 'English' : 92},
    {'name':'Sakshi', 'Science' : 55, 'Maths' : 81, 'English' : 72},
    {'name':'Naman', 'Science' : 76, 'Maths' : 88, 'English' : 94},
    {'name':'Harpreet', 'Science' : 79, 'Maths' : 72, 'English' : 89},
    ]


enter_name = input('Enter your name - ').lower()
found = False

for student in student_record:
    if student['name'].lower() == enter_name:
        found = True
        marks = (student['Science'],student['Maths'],student['English'])
        l = len(marks)
        total = sum(marks)
        avg = round(total/l)
        max_marks = max(marks)
        min_marks = min(marks)
        print('Student', '-' ,student['name'])
        print()
        print('Marks Details\n')
        print('Science', student['Science'])
        print('Maths', student['Maths'])
        print('English', student['English'])
        print('Average Marks', '-', avg)
        print('Max Marks', '-', max_marks)
        print('Min Marks', '-', min_marks)
        
if not found:
    print('Student not found in record')
```
