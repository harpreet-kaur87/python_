```python
# Write a Python program that takes a list of numbers and prints the maximum number 
# without using the built-in max() function.
```

```python
num = [12, 45, 7, 23, 89, 34]

def find_max(num):
    
    for i in range(len(num)):
        for j in range(i+1, len(num)):
            if num[i] > num[j]:
                num[i],num[j] = num[j],num[i]
    return num[-1]

answer = find_max(num)
print(answer)
```
