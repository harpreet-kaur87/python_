Student Marks Management System : 

This Python project is a simple student marks management system using a dictionary. It allows the user to add new students, update existing student marks, delete students, and display all student records. The program can also calculate the average marks and identify the student with the highest marks. This project demonstrates practical use of dictionaries, functions, and basic data management in Python.

```python
students = {
    'Alice': 85,
    'Bob': 71
}

def add_student(student,marks):
    
    if student not in students:
        students[student] = marks
        print(f'{student} has been successfully added to student record\n')
    else:
        print(f'{student} already exist in student record\n')
        
def update_marks(student,new_marks):
    
    if student in students:
        print(f'{student} marks successfully updated from {students[student]} to {new_marks}\n')
        students[student] = new_marks
    else:
        print(f'{student} not found\n')
        
def delete_student(student):
    
    if student in students:
        del students[student]
        print(f'{student} has been deleted from record\n')

    
def highest_marks(students):
    
    max_marks = max(students.values())
    result = [(k,v) for k,v in students.items() if v == max_marks]
    print(f'Student with maximum marks - {result}\n')
    
def average_marks(students):
    
    total_marks = sum(students.values())
    l = len(students)
    avg_marks = total_marks / l
    print(f'Average marks is {avg_marks}\n')
    

add_student('Damini',96)
add_student('Namit',72)
update_marks('Bob',91)
delete_student('Alice')
highest_marks(students)
average_marks(students)

    
    
print(f'Student Record- \n')

for key,value in students.items():
    print(f'{key} : {value}')
    
```
